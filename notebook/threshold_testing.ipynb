```python
alphabetical = []
for file in glob.glob('../data/11151/*.fits'):
    try:
        name = file.replace('../data/11151/', '')
        name = name.replace('.fits', '')
        e = name + f' {significant(file, threshold=0.09)}'
        alphabetical.append(e)
    except Exception:
        print('Try again for %s' % file)

alphabetical.sort()
for e in alphabetical:
    l = len(e)
    if e[l-4: l] == 'True':
        e = e[l-4: l]
    else:
        e = e[l-5: l]
    print(e)
```
